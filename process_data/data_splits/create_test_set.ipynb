```python
import os
import random
import shutil

# Define directories
original_dir = '/ess/scratch/scratch1/rachelgordon/complex_undersampled'  # Change this to your directory path
test_dir = '/ess/scratch/scratch1/rachelgordon/test_data/complex_undersampled'  # Folder where test data will be moved to

# Make sure the test directory exists
os.makedirs(test_dir, exist_ok=True)

# List all scan directories in the original directory
scan_dirs = [d for d in os.listdir(original_dir) if os.path.isdir(os.path.join(original_dir, d))]

# Randomly select 30 scan directories
selected_scans = random.sample(scan_dirs, 30)
```
