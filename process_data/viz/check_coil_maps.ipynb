```python
import numpy as np
import matplotlib.pyplot as plt

file_path = '/gpfs/data/karczmar-lab/workspaces/rachelgordon/demo_dce_recon/fastMRI_breast_002_1/coil_sensitivity_maps/coil_sensitivity_map_slice_000'

# Load the saved numpy array
mps = np.load(file_path + '.npy')

# Set up the figure and axis grid
fig, axes = plt.subplots(4, 4, figsize=(12, 12))  # 4x4 grid
fig.suptitle('Coil Sensitivity Maps - Magnitude', fontsize=16)  # Title for the whole figure

for i in range(16):
    ax = axes[i // 4, i % 4]  # Determine the subplot to use
    ax.imshow(np.squeeze(np.abs(mps[i])), cmap='gray', origin='lower')
    ax.set_title(f'Coil {i+1}')
    ax.axis('off')  # Hide axes

# Adjust layout to prevent overlapping
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Save the figure
plt.savefig(file_path + '.png', dpi=300)

plt.show()

plt.close(fig)
```

```python
# Visualize the phase

# Compute the phase of each coil sensitivity map
phase_maps = np.angle(mps)

# Set up the figure and axis grid
fig, axes = plt.subplots(4, 4, figsize=(12, 12))  # 4x4 grid
fig.suptitle('Coil Sensitivity Maps - Phase', fontsize=16)  # Title for the whole figure

for i in range(16):
    ax = axes[i // 4, i % 4]  # Determine the subplot to use
    ax.imshow(np.squeeze(phase_maps[i]), cmap='jet', origin='lower')
    ax.set_title(f'Coil {i+1}')
    ax.axis('off')  # Hide axes

# Adjust layout to prevent overlapping
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Save the figure
plt.savefig(file_path + '_phase.png', dpi=300)
plt.show()
plt.close()
```
