```python
import nibabel as nib
import matplotlib.pyplot as plt
```

```python
def load_nii(path):
    """
    Load a NIfTI image from the specified path.
    Args:
    - path (str): File path to the NIfTI image.
    """
    nii_image = nib.load(path)
    return nii_image.get_fdata()
```

```python
slice = 100
zero_pad_slice_path = f'/ess/scratch/scratch1/rachelgordon/zero_padding/fastMRI_breast_001_2/slice_{slice:03d}_frame_000.nii'
interpolated_slice_path = f'/ess/scratch/scratch1/rachelgordon/interpolation/fastMRI_breast_001_2/slice_{slice:03d}_frame_000.nii'

zero_pad_slice = load_nii(zero_pad_slice_path)
interpolated_slice = load_nii(interpolated_slice_path)
```

```python
 # Create a 1x4 plot grid
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# --- Top Row: DL Reconstruction Comparison ---
axes[0].imshow(zero_pad_slice, cmap='gray')
axes[0].set_title(f"Zero-Padded k-space Recon (Slice {slice})")

axes[1].imshow(interpolated_slice, cmap='gray')
axes[1].set_title(f"Interpolated k-space Recon (Slice {slice})")

# Turn off axes for all plots
for ax in axes.flat:
    ax.axis('off')
```
