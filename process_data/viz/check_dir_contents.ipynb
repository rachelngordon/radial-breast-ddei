```python
# check the contents of all mri scan directories to ensure the data has been processed properly

def check_directories(base_dir, expected_slices=192, expected_frames=4):
    test_set = ['fastMRI_breast_048_1', 'fastMRI_breast_114_1', 'fastMRI_breast_289_1', 'fastMRI_breast_194_1', 'fastMRI_breast_161_1', 'fastMRI_breast_288_1', 'fastMRI_breast_053_1', 'fastMRI_breast_139_1', 'fastMRI_breast_173_1', 'fastMRI_breast_179_1', 'fastMRI_breast_091_1', 'fastMRI_breast_014_1', 'fastMRI_breast_223_1', 'fastMRI_breast_285_1', 'fastMRI_breast_004_1', 'fastMRI_breast_073_1', 'fastMRI_breast_102_1', 'fastMRI_breast_096_1', 'fastMRI_breast_251_1', 'fastMRI_breast_040_1', 'fastMRI_breast_111_1', 'fastMRI_breast_108_1', 'fastMRI_breast_242_1', 'fastMRI_breast_270_1', 'fastMRI_breast_128_1', 'fastMRI_breast_050_1', 'fastMRI_breast_057_1', 'fastMRI_breast_032_1', 'fastMRI_breast_141_1']
    
    missing_dirs = []
    # List all directories in the base directory
    directories = [d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]
    
    for directory in directories:
        dir_path = os.path.join(base_dir, directory)
        
        # List all files in the current directory
        files = [f for f in os.listdir(dir_path) if os.path.isfile(os.path.join(dir_path, f))]
        
        # Check if the number of files matches the expected number
        if len(files) == expected_slices * expected_frames:
            #print(f"Directory '{directory}' contains the expected number of files ({len(files)}/{expected_slices * expected_frames}).")
            continue
        else:
            print(f"Directory '{directory}' does NOT contain the expected number of files ({len(files)}/{expected_slices * expected_frames}).")
            missing_dirs.append(directory)


    print(f"all {len(directories)} directories checked")

    return missing_dirs
```
